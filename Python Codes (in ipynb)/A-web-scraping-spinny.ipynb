```python
# Importing libraries

import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import pandas as pd
import numpy as np
```

```python
# Calling webdriver (here, the Chrome driver) using .exe file preset in local storage
path = 'C:\\Users\\R\\Desktop\\MASAI\\PROJECT\\chromedriver.exe'
browser = webdriver.Chrome(path)
```

```python
# Opening the initial web page of desired website
link = 'https://www.spinny.com/used-cars-in-delhi-ncr/s/?utm_medium=gads_c_search&utm_source=SPD-Search-Top8-National-Brand-EM-Home&utm_content=586920586020&utm_term=e_spinny&partner_id=Cj0KCQjwj_ajBhCqARIsAA37s0y4wtTE4Ilimt_HSyjbzULeMAy8Or6CCXsG_5USVxSrHmv_LcCxWY8aAh4IEALw_wcB&source=googleads&sub_source=SPD-Search-Top8-National-Brand-EM&gclid=Cj0KCQjwj_ajBhCqARIsAA37s0y4wtTE4Ilimt_HSyjbzULeMAy8Or6CCXsG_5USVxSrHmv_LcCxWY8aAh4IEALw_wcB'
browser.get(link)
```

```python
# Naming this web-page as parent
parent = browser.current_window_handle
```

```python
# Creating a list of some big cities in India
cities = ['Delhi NCR','Bangalore','Hyderabad','Mumbai','Pune','Ahmedabad','Chennai','Kolkata','Lucknow','Jaipur',
          'Chandigarh','Coimbatore','Indore','Kochi']

# Creating an empty list to hold details of cars being scraped with each innermost loop
car_info = []

# Running loop across the list - cities to access list of cars available
for city in cities:
    search = browser.find_element(By.XPATH, '//*[@id="webMainHead"]/div/div[2]/div[1]/div[1]/div/span[2]')
    search.click()
    time.sleep(1)
    
    enter_city = browser.find_element(By.XPATH, '//*[@id="webMainHead"]/div/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[1]/input')
    enter_city.send_keys(city)
    try:
        submit = browser.find_element(By.CLASS_NAME, 'styles__otherCityDivWrapper')
    except:
        submit = browser.find_element(By.CLASS_NAME, 'styles__cityText')
    submit.click()
    time.sleep(2)

    # Getting a list of all cars in particular city and setting range for iterating
    for page in range(4):
        view_button = browser.find_elements(By.CLASS_NAME, 'styles__yearAndMakeAndModelSection')
        # Entering inside each car by opeing a new tab and scraping particular details
        for i in view_button:
            print(i)
            i.click()
            time.sleep(1)
            browser.switch_to.window(browser.window_handles[-1])

            try:
                brand_and_model = browser.find_element(By.XPATH, '//*[@id="react-view"]/div[1]/div/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[1]/h1').text
            except:
                brand_and_model = np.nan
            try:
                km_driven = browser.find_element(By.XPATH,".//div[@class = 'DesktopOverview__itemDisplay DesktopOverview__textLowercase']").text
            except:
                km_drive = np.nan
            try:
                fuel_type = browser.find_element(By.XPATH, '//*[@id="react-view"]/div[1]/div/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[1]/div[2]/span[2]').text
            except:
                fuel_type = np.nan
            try:
                transmission = browser.find_element(By.XPATH, '//*[@id="react-view"]/div[1]/div/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[1]/div[2]/span[3]').text
            except:
                transmission = np.nan
            try:
                price = browser.find_element(By.XPATH, '//*[@id="react-view"]/div[1]/div/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[1]/div[5]/div[1]/span').text
            except:
                price = np.nan
            try:
                all_spec_1 = browser.find_element(By.XPATH, '//*[@id="react-view"]/div[1]/div/div[2]/div[1]/div[2]/div[1]/div[4]/section[4]/div/div[2]').text
            except:
                all_spec_1 = np.nan
            try:
                all_spec_2 = browser.find_element(By.XPATH, '//*[@id="react-view"]/div[1]/div/div[2]/div[1]/div[2]/div[1]/div[3]/section[4]/div/div[2]').text
            except:
                all_spec_2 = np.nan

            time.sleep(1)
            
            # Creating a dictionary for the values to be appended in the car_info list
            car_details = {'brand_and_model':brand_and_model, 'km_driven':km_driven, 'fuel_type': fuel_type, 'transmission':transmission,'all_spec_1':all_spec_1, 'all_spec_2': all_spec_2, 'price':price}
            car_info.append(car_details)
            
            # It's optional to extract each car's records to a csv in local storage
            # df = pd.DataFrame(car_info)
            # df.to_csv('raw.csv')
            
            # Closing this car's tab and going back to the parent tab
            browser.close()
            browser.switch_to.window(parent)
        time.sleep(3)
    time.sleep(3)
```

```python
# Printing the raw scraped dataset
df = pd.DataFrame(car_info)
df
```
